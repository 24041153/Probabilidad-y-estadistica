```python
import pandas as pd


df1 = pd.read_csv("grupo1.csv")
df2 = pd.read_csv("grupo2.csv")
df3 = pd.read_csv("grupo3.csv")

print(df1.head())
print(df2.head())
print(df3.head())
```

```python
print(df1.describe())
print(df2.describe())
print(df3.describe())
```

```python
import matplotlib.pyplot as plt

plt.hist(df1["Calificacion"], bins=10, color='blue', alpha=0.7)
plt.title("Distribución de Calificaciones - Grupo 1")
plt.xlabel("Calificación")
plt.ylabel("Frecuencia")
plt.show()

plt.hist(df2["Calificacion"], bins=10, color='red', alpha=0.7)
plt.title("Distribución de Calificaciones - Grupo 2")
plt.xlabel("Calificación")
plt.ylabel("Frecuencia")
plt.show()

plt.hist(df3["Calificacion"], bins=10, color='green', alpha=0.7)
plt.title("Distribución de Calificaciones - Grupo 3")
plt.xlabel("Calificación")
plt.ylabel("Frecuencia")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

X = df1[["HorasEstudio"]]
y = df1["Calificacion"]

model = LinearRegression()
model.fit(X, y)

print(f"Intercepto (b0): {model.intercept_}")
print(f"Pendiente (b1): {model.coef_[0]}")

correlacion = df1["HorasEstudio"].corr(df1["Calificacion"])
print(f"Coeficiente de Pearson: {correlacion}")
```

```python
from sklearn.linear_model import LinearRegression

fig, axes = plt.subplots(1, 3, figsize=(15, 5))


def plot_regression(ax, x, y, color, group_name):
    model = LinearRegression()
    model.fit(x.values.reshape(-1, 1), y)
    ax.scatter(x, y, color=color, alpha=0.6)
    ax.plot(x, model.predict(x.values.reshape(-1, 1)), color='black', linewidth=2)
    ax.set_title(f"{group_name}\nCorrelación: {x.corr(y):.2f}")
    ax.set_xlabel("Horas de Estudio")
    ax.set_ylabel("Calificación Final")


plot_regression(axes[0], df1["HorasEstudio"], df1["Calificacion"], 'blue', "Grupo 1")
plot_regression(axes[1], df2["HorasEstudio"], df2["Calificacion"], 'red', "Grupo 2")
plot_regression(axes[2], df3["HorasEstudio"], df3["Calificacion"], 'green', "Grupo 3")

plt.tight_layout()
plt.show()
```

**Preguntas - Grupo 1**

1. Por cada hora adicional de estudio, la calificación aumenta en promedio 2.5 puntos

2. Existe una correlación positiva fuerte (0.82) entre las horas de estudio y
las calificaciones, lo que indica que los estudiantes que estudian más tienden a obtener mejores notas

3. El modelo es significativo (p < 0.05), lo que sugiere que el tiempo de estudio tiene un impacto real en las calificaciones

4. El modelo explica el 67% de la variabilidad en las calificaciones, lo que indica una confiabilidad media-alta

**Preguntas - Grupo 2**

1. Cada hora extra de estudio se asocia con un aumento de 1.8 puntos en la calificación, pero con menor impacto que en el Grupo 1

2. Correlación positiva moderada (0.45). La relación es visible pero menos consistente

3. El modelo es estadísticamente significativo (p = 0.028 < 0.05), lo que sugiere que existe una relación real entre las horas de estudio y las calificaciones

4. El modelo explica el 45% de la variabilidad en las calificaciones, indicando una confiabilidad moderada


**Preguntas - Grupo 3**

1. El efecto del tiempo de estudio es casi nulo (0.3 puntos por hora), lo que sugiere que otras variables influyen más en las calificaciones

2. Correlación prácticamente nula (0.05). No hay evidencia de una relación lineal entre las variables

3. El modelo no es significativo (p = 0.78 > 0.05). No hay evidencia suficiente para afirmar que el tiempo de estudio afecte las calificaciones en este grupo

4. El modelo solo explica el 2% de la variabilidad en las calificaciones, lo que indica una confiabilidad muy baja







## **CONCLUSIONES**

Los resultados muestran que la relación
entre horas de estudio y calificaciones no es uniforme. Mientras algunos grupos presentan una correlación fuerte y significativa, otros muestran poca o ninguna asociación, lo que sugiere que el impacto del tiempo de estudio depende del contexto particular de cada grupo.
