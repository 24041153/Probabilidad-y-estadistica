```python
import math as mt
from itertools import permutations
from itertools import product
import numpy as np
```

```python
Asistentes = {"A", "B", "C", "D", "E", "F", "G", "H", "I", "J"}
n = len(Asistentes)
r = 4

Cn = mt.factorial(n) / (mt.factorial(n-r))
print("Existen", Cn, "posibles combinaciones sin repeticion del total de",
       n , " asistentes ", "en grupos de ", r, "en" , r)

print("Aqui las permutaciones")
permutaciones_asistentes = list(permutations(Asistentes, r))

for permutacion in permutaciones_asistentes:
  print(permutacion)
```
