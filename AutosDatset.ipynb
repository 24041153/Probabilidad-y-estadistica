```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
from statistics import mode
```

```python
autos = pd.read_csv("autos2.csv", encoding = "ISO-8859-1")
```

```python
autos.head()
```

```python
autos.info()
```

```python
Minimo = autos.price.min()
Maximo = autos.price.max()
Moda = mode(autos.price)
Mediana = autos.price.median()
Media = autos.price.mean()
Desviacion = autos.price.std()

print("Minimo =", Minimo)
print("Maximo =", Maximo)
print("Moda =", Moda)
print("Mediana = ", Mediana)
print("Media=", Media )
print("Desviacion =", Desviacion)

```

```python
autos.describe()
```

```python
import math

n = len(autos.price)
k = int( n + math.log2(10))
print("Valor n = ", n , "\nNumero de intervalos =" , k)
```

```python
media = autos.price.mean()

plt.pyplot.hist(autos.price, bins =20)
plt.pyplot.title("Histograma de precios Autodata Set")
plt.pyplot.axvline(media, color="y", linestyle="dashed", linewidth=2, label=f"Media={media:.2f}")
plt.pyplot.xlabel("Precio")
plt.pyplot.ylabel("Fruecuencia")

# En el histograma se muestran las veces que se repite un valor en especifico
```

```python
sns.boxplot(x="body-style", y="price", data=autos)
plt.title("Distribuci√≥n de precios por tipo de vehiculo")
plt.show()

# Muestra la distribuci√≥n de precios, con medianas y cuartiles
```

```python
sns.boxplot(x='drive-wheels', y='price', data=autos)
plt.title("Distribuci√≥n de precios por tipo de tracci√≥n")
plt.show()

# Visualizacion diferencias de precios seg√∫n el sistema de tracci√≥n de los autos
```

```python
sns.boxplot(x='engine-location', y="price", data=autos)
plt.title("Relacion entre el precio y la localizacion del motor ")
plt.show()

# Comparacion entre la ubicacion de el motor (En frente y Hacia Atras) y la frecuencia de precios
```
